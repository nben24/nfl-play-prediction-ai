```python
import pandas as pd

url = "https://github.com/nflverse/nflverse-data/releases/download/pbp/play_by_play_2025.csv"

cols = [
    "play_type",
    "down",
    "ydstogo",
    "yardline_100",
    "qtr",
    "score_differential",
    "game_seconds_remaining"
]

df = pd.read_csv(url, usecols=cols)

# keep only run/pass
df = df[df["play_type"].isin(["run", "pass"])]

# sample for speed
df = df.sample(10000, random_state=42)

print(df.head())
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
```
